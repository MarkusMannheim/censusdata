```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns
import geopandas as gpd, topojson as tp

sns.set(font_scale=1.25)

data_2006 = pd.read_csv(
    "publicHousing2006.csv",
    skiprows=11,
    skipfooter=10,
    usecols=[0, 2, 8, 11],
    names=["suburb", "public", "community", "total"],
    index_col="suburb",
    engine="python"
)

data_2011 = pd.read_csv(
    "publicHousing2011.csv",
    skiprows=11,
    skipfooter=11,
    usecols=[0, 2, 8, 11],
    names=["suburb", "public", "community", "total"],
    index_col="suburb",
    engine="python"
)

data_2016 = pd.read_csv(
    "publicHousing2016.csv",
    skiprows=11,
    skipfooter=10,
    usecols=[0, 2, 8, 11],
    names=["suburb", "public", "community", "total"],
    index_col="suburb",
    engine="python"
)

data_2021 = pd.read_csv(
    "publicHousing2021.csv",
    skiprows=11,
    skipfooter=10,
    usecols=[0, 2, 3, 11],
    names=["suburb", "public", "community", "total"],
    index_col="suburb",
    engine="python"
)

housing = pd.DataFrame(
    columns=["year_2006", "year_2011", "year_2016", "year_2021"]
)
housing.index.name = "suburb"

for suburb in data_2006.index:
    housing.at[suburb.replace(" - ACT", ""), ["year_2006"]] = data_2006.at[suburb, "public"] / data_2006.at[suburb, "total"]

for suburb in data_2011.index:
    housing.at[suburb.replace(" (ACT)", ""), ["year_2011"]] = data_2011.at[suburb, "public"] / data_2011.at[suburb, "total"]

for suburb in data_2016.index:
    if "ACT Remainder" not in suburb:
        housing.at[suburb.replace(" (ACT)", ""), ["year_2016"]] = data_2016.at[suburb, "public"] / data_2016.at[suburb, "total"]

for suburb in data_2021.index:
    if "ACT Remainder" not in suburb:
        housing.at[suburb.replace(" (ACT)", ""), ["year_2021"]] = data_2021.at[suburb, "public"] / data_2021.at[suburb, "total"]

(housing * 100).to_csv("./public_housing.csv")
```

```python
years = ["year_2006", "year_2011", "year_2016", "year_2021"]

fig, ax = plt.subplots(
    4, 1,
    figsize=(8, 18),
    tight_layout=True,    
    sharex=True
)

for i, year in enumerate(years):    
    sns.kdeplot(
        data=housing[year],
        ax=ax[i],
        color=sns.color_palette()[i],
        linewidth=3
    )    
    ax[i].vlines(
        x=housing[year].median(),
        ymin=0,
        ymax=ax[i].get_ylim()[1],
        ls='--',
        lw=2,
        color=sns.color_palette()[i]
    )    
    ax[i].xaxis.set_major_formatter(lambda x, p: f"{x:.0%}")
    ax[i].xaxis.set_label_text(None)
    ax[i].yaxis.set_label_text(None)
    ax[i].set_yticklabels([])
    ax[i].set_title(year[5:], fontsize="x-large")
    
    print(housing[year].mean())
    
plt.xlim([0, .3])
plt.suptitle("Public housing (% of dwellings per suburb)", fontsize="x-large", fontweight="bold")
plt.savefig("./publicHousing.jpg")
plt.show()
```

```python
sals = gpd.read_file(
    "./sals_2021.zip"
).iloc[:, [1, 2, 6, 10]]
sals.columns = ["name", "state", "area", "geometry"]
sals = sals[
    (sals["area"] > 0) &
    (sals["state"] == "8") &
    (sals["name"].apply(lambda x: x if "ACT Remainder" not in x else False))
]
sals["name"] = sals["name"].str.replace(" (ACT)", "", regex=False)
sals.set_index("name", inplace=True)

for suburb in housing.index:
    for year in years:
        sals.at[suburb, year] = housing.at[suburb, year]

sals = sals[years + ["geometry"]]
sals.to_file("./publicHousing.geojson")
```

```python
fig, ax = plt.subplots(
    4, 1,
    figsize=(10, 40),
    tight_layout=True,
    sharex=True
)

for i, year in enumerate(years):    
    sals.plot(
        year,
        cmap="mako_r",
        ax=ax[i],
        legend=True,
        legend_kwds={
            "location":"top",
            "shrink":.5
        }
    )
    ax[i].set_axis_off()
```

```python
topo = tp.Topology(
    sals.fillna(value="null").reset_index(),
    toposimplify=1e-4
).topoquantize(1e4)
topo.to_json("./publicHousing.json")
```
