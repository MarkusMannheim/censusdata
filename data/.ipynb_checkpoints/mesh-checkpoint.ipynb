```python
import pandas as pd
import numpy as np
import geopandas as gpd

data = pd.DataFrame(
    columns=["code", "category", "area", "dwellings", "people", "state"]
)

for sheet in range(1, 13):
    sheet_data = pd.read_excel(
        "./MeshBlockCounts2021.xlsx",
        sheet_name=sheet,
        header=6,
        skipfooter=4
    )
    sheet_data.columns = data.columns
    data = pd.concat([data, sheet_data])
    print("sheet", sheet, "completed; length now", len(data))
```

```python
test = data.copy()
test["dwellings"] = test["dwellings"].astype("float")
test["people"] = test["people"].astype("float")
test = test[
    (test["code"].apply(lambda x: str(x)[0] == "8")) &
    (test["area"] > 0)
]
test.groupby("category").sum()
# test.dtypes
```
