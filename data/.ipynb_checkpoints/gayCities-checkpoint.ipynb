```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import seaborn as sns
sns.set(
    style="whitegrid",
    font_scale=1.25
)

suburb_data = pd.read_csv(
    "./gayburbs.csv",
    index_col="code"
)

suburb_geodata = gpd.read_file(
    "./SAL_2021_AUST_GDA2020_SHP.zip"
).iloc[:, [0, 6, 10]]
suburb_geodata.columns = ["code", "area", "geometry"]
suburb_geodata.set_index("code", inplace=True)
suburb_geodata = suburb_geodata[suburb_geodata["area"] > 0]

city_geodata = gpd.read_file(
    "./SA4_2021_AUST_SHP_GDA2020.zip"
).iloc[:, [0, 1, 10, 12]]
city_geodata.columns = ["code", "name", "area", "geometry"]
city_geodata.set_index("code", inplace=True)
city_geodata = city_geodata[city_geodata["area"] > 0]
```

```python
sa4s = pd.DataFrame(columns=["name", "area"])
sa4s.index.name = "code"

lines = [
    Line2D([0], [0], color=sns.color_palette()[0], lw=4),
    Line2D([0], [0], color=sns.color_palette()[2], lw=4)
]

for city in city_geodata.index:
    print("analysing", city_geodata.at[city, "name"], "...")
    city_shape = city_geodata.at[city, "geometry"].buffer(.005)
    suburbs = suburb_geodata[suburb_geodata["geometry"].within(city_shape)]
    data = suburb_data.loc[["SAL" + x for x in suburbs.index.tolist()]].loc[:, "male couples":].sum()
    sa4s.at[city, "name"] = city_geodata.at[city, "name"]
    sa4s.at[city, "area"] = city_geodata.at[city, "area"]
    for measure in data.index:
        sa4s.at[city, measure] = data[measure]
    
sa4s["gay"] = sa4s["male couples"] / sa4s["total couples"]
sa4s["lesbian"] = sa4s["female couples"] / sa4s["total couples"]
sa4s["same-sex"] = sa4s["gay"] + sa4s["lesbian"]
sa4s = sa4s.loc[:, ["name", "gay", "lesbian", "same-sex"]]
```

```python
sa4s = sa4s.loc[:, ["name", "gay", "lesbian", "same-sex"]].sort_values("same-sex", ascending=False)
sa4s.to_csv("./gayCities.csv")
sa4s
```
