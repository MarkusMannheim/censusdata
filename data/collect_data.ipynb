```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
from os import listdir
from os.path import isfile, join
sns.set(style="whitegrid", font_scale=1.25)

names = pd.read_excel(
    "./root/metadata/2021Census_geog_desc_1st_release.xlsx",
    sheet_name=3,
    index_col="Census_Code_2021"
)

data = pd.read_csv(
    "./root/data/SAL/AUS/2021Census_G33_AUST_SAL.csv",
    index_col="SAL_CODE_2021",
    usecols=(lambda x: (("_Tot" in x) & ("stated" not in x) & (x != "Tot_Tot")) | (x == "SAL_CODE_2021"))
)

midpoints = [0.0]

for col in data.columns[1:]:
    lower = col.split("_")[1]
    upper = col.split("_")[2]
    if upper == "more":
        upper = 4499.0
    midpoints.append((float(lower) + float(upper)) / 2)

def find_median(indice):
    values = data.loc[indice, :"HI_4000_more_Tot"]
    list_list = pd.Series([([midpoints[i]] * int(n)) for i, n in enumerate(values)])
    median = pd.Series(itertools.chain.from_iterable(list_list)).median()
    if pd.notna(median):
        data.at[indice, "median"] = median

for i in data.index:
    find_median(i)
    data.at[i, "area"] = names.at[i, "Census_Name_2021"]
```

```python
fig = plt.subplots(
    figsize=(12, 6.75),
    tight_layout=True
)
sns.histplot(
    data=data["median"],    
    bins=17,
    kde=True,
    line_kws=dict(linewidth=3)
)
plt.margins(x=0)
plt.title("Distribution of median household incomes (localities)", fontweight="bold", fontsize="large")
plt.xlabel(None)
plt.ylabel(None)
```
